```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Market_basket.csv',header=None)
df.head(5)
```

```python
df.shape
```

```python
transactions = []
for i in range(0,7501):
    transactions.append([str(df.values[i,j]) for j in range(0,20)])
transactions
```

```python
# Training the appriori model on dataset
```

```python
from apyori import apriori
rules = apriori(transactions=transactions, min_support=0.003, 
                min_confidence=0.2,
                min_lift=3, 
                min_length=2,
                  max_length=2)
rules
```

```python
result = list(rules)
```

```python
len(result)
```

```python
def inspect(result):
    lhs = [tuple(result[2][0][0])[0] for result in result]
    rhs = [tuple(result[2][0][1])[0] for result in result]
    supports = [result[1] for result in result]
    confidence = [result[2][0][2] for result in result]
    lifts = [result[2][0][3] for result in result]
    return list(zip(lhs,rhs,supports,confidence,lifts))
resultindataframe = pd.DataFrame(inspect(result), columns = ['LHS', 'RHS', 'Support', 'Confidence', 'Lift'])
resultindataframe
```
